```python
# DO NOT MODIFY THIS CELL

from abc import ABC, abstractmethod  

class AbstractSearchInterface(ABC):
    '''
    Abstract class to support search/insert operations (plus underlying data structure)
    
    '''
        
    @abstractmethod
    def insertElement(self, element):     
        '''
        Insert an element in a search tree
            Parameters:
                    element: string to be inserted in the search tree (string)

            Returns:
                    "True" after successful insertion, "False" if element is already present (bool)
        '''
        
        pass 
    

    @abstractmethod
    def searchElement(self, element):
        '''
        Search for an element in a search tree
            Parameters:
                    element: string to be searched in the search tree (string)

            Returns:
                    "True" if element is found, "False" otherwise (bool)
        '''

        pass
```

```python
# TWO-THREE TREE: Auxillary data structures (node class) and helper functions/methods:

class TwoThreeNode:
    def __init__(self, data, parent = None):
        self.data = list([data])
        self.parent = parent
        self.child = list()
    
    # Function to check if value is in tree
    def get(self, value):
        if value in self.data:
            return value
        elif self.is_leaf(): 
            return None
        else:
            return self.get_appropriate_child(value).get(value)
        
    
    # Function to check if current node is a leaf node
    def is_leaf(self):
        return len(self.child) == 0
    
    # Function to find the appropriate child leaf for any given value
    def get_appropriate_child(self, value):
        
        # Iterate through the sorted data values
        for i in range(len(self.data)):
            if value < self.data[i]:
                # The value belongs in the child to the left of this data point
                return self.child[i]
            
        # If the value is greater than all data points, it belongs to the last child
        return self.child[-1]

    # Function to insert a new node into the tree
    def insert(self, new_node):
        if self.is_leaf():
            self.add(new_node)
        
        #if not leaf node - find correct child and call insert method again
        else:
            child = self.get_appropriate_child(new_node.data[0])
            child.insert(new_node)

    # Function to add a new node and reorganises tree 
    def add(self, new_node):
        for child in new_node.child:
            child.parent = self
        self.data.extend(new_node.data)
        self.data.sort()
        self.child.extend(new_node.child)
        if len(self.child) > 1:
            for child in self.child:
                child.parent = self
        if len(self.data) > 2:
            self.split()
    
    # If there are more than 2 data items in a node, we need to split into a new subtree and promote the middle
    def split(self):
        # Create left and right children of new subtree and set the new parent as the current "self" node
        left_child = TwoThreeNode(self.data[0], self)
        right_child = TwoThreeNode(self.data[2], self)

        # Reassign children of initial self node to the left and right children of the new nodes we created
        if self.child:
            self.child[0].parent = left_child
            self.child[1].parent = left_child
            self.child[2].parent = right_child
            self.child[3].parent = right_child
            left_child.child = [self.child[0], self.child[1]]
            right_child.child = [self.child[2], self.child[3]]

        # Assign left and right child of new parent node as the inital left and right children
        # Set data of parent node to only contain data from the inital middle child
        self.child = [left_child]
        self.child.append(right_child)
        self.data = [self.data[1]]
        
        # Promote parent node
        if self.parent:
            if self in self.parent.child:
                self.parent.child.remove(self)
            self.parent.add(self)
        else:
            left_child.parent = self
            right_child.parent = self
        
    def remove(self, item):
        pass
    
```

```python
# LLRB: Auxillary Data Structures

RED = True
BLACK = False


class LLRBNode:
    def __init__(self, key, color=RED):
        self.key = key
        self.color = color
        self.left = None
        self.right = None


def _is_red(node):
    return node is not None and node.color == RED
```

```python
# AVL Tree: Auxillary Data Structures & Helper functions

class AVLNode:
    def __init__(self, element): 
        self.element = element
        self.left = None
        self.right = None
        self.height = 1 # height subtree with node as root

def get_height(node):
    return node.height if node else 0

def get_balance(node):
    if not node:
        return 0
    return get_height(node.left) - get_height(node.right)

# Updates height of node, called upon insertion
def update_height(node):
    return 1 + max(get_height(node.left), get_height(node.right))

# Right rotation, makes the left node of the old root the new root, returns new root
def rotate_right(old_root):
    new_root = old_root.left
    subtree = new_root.right
    new_root.right = old_root
    old_root.left = subtree
    old_root.height = update_height(old_root)
    new_root.height = update_height(new_root)
    return new_root

# Left rotation, makes the right node of the old root the new root, returns new root
def rotate_left(old_root):
    new_root = old_root.right
    subtree = new_root.left
    new_root.left = old_root
    old_root.right = subtree
    old_root.height = update_height(old_root)
    new_root.height = update_height(new_root)
    return new_root

# Checks for imbalance, and determines which case and applies corresponding rotations to fix imbalance
# 4 cases: LL, RR, LR, RL (links)
def rebalance(node, element):
    balance = get_balance(node)
    
    # Left Heavy
    if balance > 1:
        if element < node.left.element: # LL
            return rotate_right(node)
        else: # LR
            node.left = rotate_left(node.left)
            return rotate_right(node)
            
    # Right Heavy
    if balance < -1:
        if element > node.right.element: # RR
            return rotate_left(node)
        else: # RL
            node.right = rotate_right(node.right)
            return rotate_left(node)
            
    return node
```

```python
class AVLTree(AbstractSearchInterface):
    def __init__(self):
        self.root = None

    def insertElement(self, element):
        # We call the recursive helper starting at the root
        self.root = self._insert_recursive(self.root, element)

    def _insert_recursive(self, node, element):
        if node is None:
            return AVLNode(element)
        
        if element < node.element:
            node.left = self._insert_recursive(node.left, element)
        elif element > node.element:
            node.right = self._insert_recursive(node.right, element)
        else:
            return node

        node.height = update_height(node)
        
        return rebalance(node, element)

    def searchElement(self, element):
        current = self.root
        while current:
            if element == current.element:
                return True
            current = current.left if element < current.element else current.right
        return False
```

```python
class LLRBBST(AbstractSearchInterface):
    def __init__(self):
        self.root = None

    def _rotate_left(self, h):
        x = h.right
        h.right = x.left
        x.left = h
        x.color = h.color
        h.color = RED
        return x

    def _rotate_right(self, h):
        x = h.left
        h.left = x.right
        x.right = h
        x.color = h.color
        h.color = RED
        return x

    def _flip_colors(self, h):
        h.color = RED
        h.left.color = BLACK
        h.right.color = BLACK

    def _insert(self, h, element):
        if h is None:
            return LLRBNode(element, RED), True

        inserted = False
        if element < h.key:
            h.left, inserted = self._insert(h.left, element)
        elif element > h.key:
            h.right, inserted = self._insert(h.right, element)
        else:
            return h, False

        if _is_red(h.right) and not _is_red(h.left):
            h = self._rotate_left(h)
        if _is_red(h.left) and _is_red(h.left.left):
            h = self._rotate_right(h)
        if _is_red(h.left) and _is_red(h.right):
            self._flip_colors(h)

        return h, inserted

    def insertElement(self, element):
        inserted = False

        self.root, inserted = self._insert(self.root, element)
        if self.root is not None:
            self.root.color = BLACK

        return inserted

    def searchElement(self, element):
        found = False

        current = self.root
        while current is not None:
            if element < current.key:
                current = current.left
            elif element > current.key:
                current = current.right
            else:
                found = True
                break

        return found

```

```python
import random
import string

class TestDataGenerator():
    '''
    Synthetic data generator for string datasets.
    '''

    LENGTH_MIN = 20
    LENGTH_MAX = 25

    def __init__(self):
        pass

    def random_string(self):
        length = random.randint(self.LENGTH_MIN, self.LENGTH_MAX)
        alphabet = string.ascii_letters + string.digits
        return ''.join(random.choices(alphabet, k=length))

    def random_dataset(self, count):
        dataset = []
        seen = set()
        while len(dataset) < count:
            value = self.random_string()
            if value not in seen:
                seen.add(value)
                dataset.append(value)
        return dataset

    def sorted_dataset(self, count):
        return sorted(self.random_dataset(count))

    def reverse_sorted_dataset(self, count):
        return sorted(self.random_dataset(count), reverse=True)

    def make_datasets(self, count):
        return {
            'random': self.random_dataset(count),
            'sorted': self.sorted_dataset(count),
            'reverse_sorted': self.reverse_sorted_dataset(count)
        }

```

Use the cell below to implement the requested **experimental framework** (be mindful of code readability and reusability).

```python
import timeit
import matplotlib.pyplot as plt

class ExperimentalFramework():
    '''
    Experimental framework for insert and search(hit) benchmarking.
    '''

    DATASET_TYPES = ('random', 'sorted', 'reverse_sorted')

    def __init__(self, generator, tree_factories, sizes, repeats=5, query_count=None):
        self.generator = generator
        self.tree_factories = tree_factories
        self.sizes = sizes
        self.repeats = repeats
        self.query_count = query_count

    def _median(self, values):
        ordered = sorted(values)
        n = len(ordered)
        mid = n // 2
        if n % 2 == 1:
            return ordered[mid]
        return (ordered[mid - 1] + ordered[mid]) / 2

    def _time_insert(self, tree_factory, dataset):
        tree = tree_factory()
        start = timeit.default_timer()
        for element in dataset:
            tree.insertElement(element)
        elapsed = timeit.default_timer() - start
        return elapsed

    def _build_tree(self, tree_factory, dataset):
        tree = tree_factory()
        for element in dataset:
            tree.insertElement(element)
        return tree

    def _build_hit_queries(self, dataset):
        if not dataset:
            return []
        n = self.query_count if self.query_count is not None else len(dataset)
        queries = []
        for _ in range(n):
            queries.append(random.choice(dataset))
        return queries

    def _time_search_hit(self, tree_factory, dataset):
        tree = self._build_tree(tree_factory, dataset)
        queries = self._build_hit_queries(dataset)

        start = timeit.default_timer()
        for q in queries:
            tree.searchElement(q)
        elapsed = timeit.default_timer() - start
        return elapsed

    def run(self):
        results = {}
        for tree_name, tree_factory in self.tree_factories.items():
            results[tree_name] = {
                'insert': {dtype: [] for dtype in self.DATASET_TYPES},
                'search_hit': {dtype: [] for dtype in self.DATASET_TYPES}
            }

            for size in self.sizes:
                datasets = self.generator.make_datasets(size)

                for dtype in self.DATASET_TYPES:
                    insert_runs = []
                    search_runs = []

                    for _ in range(self.repeats):
                        dataset = list(datasets[dtype])
                        insert_runs.append(self._time_insert(tree_factory, dataset))
                        search_runs.append(self._time_search_hit(tree_factory, dataset))

                    results[tree_name]['insert'][dtype].append(self._median(insert_runs))
                    results[tree_name]['search_hit'][dtype].append(self._median(search_runs))

        return results

    def plot_per_tree(self, results):
        for tree_name in self.tree_factories.keys():
            plt.figure(figsize=(12, 5))

            plt.subplot(1, 2, 1)
            for dtype in self.DATASET_TYPES:
                plt.plot(self.sizes, results[tree_name]['insert'][dtype], marker='o', label=dtype)
            plt.title(tree_name + ' - Insert')
            plt.xlabel('Dataset Size')
            plt.ylabel('Time (seconds)')
            plt.legend()
            plt.grid(True)

            plt.subplot(1, 2, 2)
            for dtype in self.DATASET_TYPES:
                plt.plot(self.sizes, results[tree_name]['search_hit'][dtype], marker='o', label=dtype)
            plt.title(tree_name + ' - Search (Hit)')
            plt.xlabel('Dataset Size')
            plt.ylabel('Time (seconds)')
            plt.legend()
            plt.grid(True)

            plt.tight_layout()
            plt.show()

    def plot_random_comparison(self, results):
        plt.figure(figsize=(8, 5))
        for tree_name in self.tree_factories.keys():
            plt.plot(self.sizes, results[tree_name]['insert']['random'], marker='o', label=tree_name)
        plt.title('Random Dataset - Insert Comparison (All Trees)')
        plt.xlabel('Dataset Size')
        plt.ylabel('Time (seconds)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(8, 5))
        for tree_name in self.tree_factories.keys():
            plt.plot(self.sizes, results[tree_name]['search_hit']['random'], marker='o', label=tree_name)
        plt.title('Random Dataset - Search(Hit) Comparison (All Trees)')
        plt.xlabel('Dataset Size')
        plt.ylabel('Time (seconds)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

```
