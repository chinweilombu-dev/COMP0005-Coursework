```python
# TWO-THREE TREE: Auxillary data structures (node class) and helper functions/methods:

class TwoThreeNode:
    def __init__(self, data, parent = None):
        self.data = list([data])
        self.parent = parent
        self.child = list()
    
    # Function to check if value is in tree
    def get(self, value):
        if value in self.data:
            return value
        elif self.is_leaf(): 
            return None
        else:
            return self.get_appropriate_child(value).get(value)
        
    
    # Function to check if current node is a leaf node
    def is_leaf(self):
        return len(self.child) == 0
    
    # Function to find the appropriate child leaf for any given value
    def get_appropriate_child(self, value):
        
        # Iterate through the sorted data values
        for i in range(len(self.data)):
            if value < self.data[i]:
                # The value belongs in the child to the left of this data point
                return self.child[i]
            
        # If the value is greater than all data points, it belongs to the last child
        return self.child[-1]

    # Function to insert a new node into the tree
    def insert(self, new_node):
        if self.is_leaf():
            self.add(new_node)
        
        #if not leaf node - find correct child and call insert method again
        else:
            child = self.get_appropriate_child(new_node.data[0])
            child.insert(new_node)

    # Function to add a new node and reorganises tree 
    def add(self, new_node):
        for child in new_node.child:
            child.parent = self
        self.data.extend(new_node.data)
        self.data.sort()
        self.child.extend(new_node.child)
        if len(self.child) > 1:
            for child in self.child:
                child.parent = self
        if len(self.data) > 2:
            self.split()
    
    # If there are more than 2 data items in a node, we need to split into a new subtree and promote the middle
    def split(self):
        # Create left and right children of new subtree and set the new parent as the current "self" node
        left_child = TwoThreeNode(self.data[0], self)
        right_child = TwoThreeNode(self.data[2], self)

        # Reassign children of initial self node to the left and right children of the new nodes we created
        if self.child:
            self.child[0].parent = left_child
            self.child[1].parent = left_child
            self.child[2].parent = right_child
            self.child[3].parent = right_child
            left_child.child = [self.child[0], self.child[1]]
            right_child.child = [self.child[2], self.child[3]]

        # Assign left and right child of new parent node as the inital left and right children
        # Set data of parent node to only contain data from the inital middle child
        self.child = [left_child]
        self.child.append(right_child)
        self.data = [self.data[1]]
        
        # Promote parent node
        if self.parent:
            if self in self.parent.child:
                self.parent.child.remove(self)
            self.parent.add(self)
        else:
            left_child.parent = self
            right_child.parent = self
        
    def remove(self, item):
        pass
    
```

```python
class LLRBBST(AbstractSearchInterface):
    def __init__(self):
        self.root = None

    def _rotate_left(self, h):
        x = h.right
        h.right = x.left
        x.left = h
        x.color = h.color
        h.color = RED
        return x

    def _rotate_right(self, h):
        x = h.left
        h.left = x.right
        x.right = h
        x.color = h.color
        h.color = RED
        return x

    def _flip_colors(self, h):
        h.color = RED
        h.left.color = BLACK
        h.right.color = BLACK

    def _insert(self, h, element):
        if h is None:
            return LLRBNode(element, RED), True

        inserted = False
        if element < h.key:
            h.left, inserted = self._insert(h.left, element)
        elif element > h.key:
            h.right, inserted = self._insert(h.right, element)
        else:
            return h, False

        if _is_red(h.right) and not _is_red(h.left):
            h = self._rotate_left(h)
        if _is_red(h.left) and _is_red(h.left.left):
            h = self._rotate_right(h)
        if _is_red(h.left) and _is_red(h.right):
            self._flip_colors(h)

        return h, inserted

    def insertElement(self, element):
        inserted = False

        self.root, inserted = self._insert(self.root, element)
        if self.root is not None:
            self.root.color = BLACK

        return inserted

    def searchElement(self, element):
        found = False

        current = self.root
        while current is not None:
            if element < current.key:
                current = current.left
            elif element > current.key:
                current = current.right
            else:
                found = True
                break

        return found

```
