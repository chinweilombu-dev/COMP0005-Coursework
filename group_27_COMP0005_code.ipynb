```python
# TWO-THREE TREE: Auxillary data structures (node class) and helper functions/methods:

class TwoThreeNode:
    def __init__(self, data, parent = None):
        self.data = list([data])
        self.parent = parent
        self.child = list()
    
    # Function to check if value is in tree
    def get(self, value):
        if value in self.data:
            return value
        elif self.is_leaf(): 
            return None
        else:
            return self.get_appropriate_child(value).get(value)
        
    
    # Function to check if current node is a leaf node
    def is_leaf(self):
        return len(self.child) == 0
    
    # Function to find the appropriate child leaf for any given value
    def get_appropriate_child(self, value):
        
        # Iterate through the sorted data values
        for i in range(len(self.data)):
            if value < self.data[i]:
                # The value belongs in the child to the left of this data point
                return self.child[i]
            
        # If the value is greater than all data points, it belongs to the last child
        return self.child[-1]

    # Function to insert a new node into the tree
    def insert(self, new_node):
        if self.is_leaf():
            self.add(new_node)
        
        #if not leaf node - find correct child and call insert method again
        else:
            child = self.get_appropriate_child(new_node.data[0])
            child.insert(new_node)

    # Function to add a new node and reorganises tree 
    def add(self, new_node):
        for child in new_node.child:
            child.parent = self
        self.data.extend(new_node.data)
        self.data.sort()
        self.child.extend(new_node.child)
        if len(self.child) > 1:
            for child in self.child:
                child.parent = self
        if len(self.data) > 2:
            self.split()
    
    # If there are more than 2 data items in a node, we need to split into a new subtree and promote the middle
    def split(self):
        # Create left and right children of new subtree and set the new parent as the current "self" node
        left_child = TwoThreeNode(self.data[0], self)
        right_child = TwoThreeNode(self.data[2], self)

        # Reassign children of initial self node to the left and right children of the new nodes we created
        if self.child:
            self.child[0].parent = left_child
            self.child[1].parent = left_child
            self.child[2].parent = right_child
            self.child[3].parent = right_child
            left_child.child = [self.child[0], self.child[1]]
            right_child.child = [self.child[2], self.child[3]]

        # Assign left and right child of new parent node as the inital left and right children
        # Set data of parent node to only contain data from the inital middle child
        self.child = [left_child]
        self.child.append(right_child)
        self.data = [self.data[1]]
        
        # Promote parent node
        if self.parent:
            if self in self.parent.child:
                self.parent.child.remove(self)
            self.parent.add(self)
        else:
            left_child.parent = self
            right_child.parent = self
        
    def remove(self, item):
        pass
    
```

```python
# LLRB: Auxillary Data Structures

RED = True
BLACK = False


class LLRBNode:
    def __init__(self, key, color=RED):
        self.key = key
        self.color = color
        self.left = None
        self.right = None


def _is_red(node):
    return node is not None and node.color == RED
```

```python
# AVL Tree: Auxillary Data Structures & Helper functions

class AVLNode:
    def __init__(self, element): 
        self.element = element
        self.left = None
        self.right = None
        self.height = 1 # height subtree with node as root

def get_height(node):
    return node.height if node else 0

def get_balance(node):
    if not node:
        return 0
    return get_height(node.left) - get_height(node.right)

# Updates height of node, called upon insertion
def update_height(node):
    return 1 + max(get_height(node.left), get_height(node.right))

# Right rotation, makes the left node of the old root the new root, returns new root
def rotate_right(old_root):
    new_root = old_root.left
    subtree = new_root.right
    new_root.right = old_root
    old_root.left = subtree
    old_root.height = update_height(old_root)
    new_root.height = update_height(new_root)
    return new_root

# Left rotation, makes the right node of the old root the new root, returns new root
def rotate_left(old_root):
    new_root = old_root.right
    subtree = new_root.left
    new_root.left = old_root
    old_root.right = subtree
    old_root.height = update_height(old_root)
    new_root.height = update_height(new_root)
    return new_root

# Checks for imbalance, and determines which case and applies corresponding rotations to fix imbalance
# 4 cases: LL, RR, LR, RL (links)
def rebalance(node, element):
    balance = get_balance(node)
    
    # Left Heavy
    if balance > 1:
        if element < node.left.element: # LL
            return rotate_right(node)
        else: # LR
            node.left = rotate_left(node.left)
            return rotate_right(node)
            
    # Right Heavy
    if balance < -1:
        if element > node.right.element: # RR
            return rotate_left(node)
        else: # RL
            node.right = rotate_right(node.right)
            return rotate_left(node)
            
    return node
```

```python
class AVLTree(AbstractSearchInterface):
    def __init__(self):
        self.root = None

    def insertElement(self, element):
        # We call the recursive helper starting at the root
        self.root = self._insert_recursive(self.root, element)

    def _insert_recursive(self, node, element):
        if node is None:
            return AVLNode(element)
        
        if element < node.element:
            node.left = self._insert_recursive(node.left, element)
        elif element > node.element:
            node.right = self._insert_recursive(node.right, element)
        else:
            return node

        node.height = update_height(node)
        
        return rebalance(node, element)

    def searchElement(self, element):
        current = self.root
        while current:
            if element == current.element:
                return True
            current = current.left if element < current.element else current.right
        return False
```

```python
class LLRBBST(AbstractSearchInterface):
    def __init__(self):
        self.root = None

    def _rotate_left(self, h):
        x = h.right
        h.right = x.left
        x.left = h
        x.color = h.color
        h.color = RED
        return x

    def _rotate_right(self, h):
        x = h.left
        h.left = x.right
        x.right = h
        x.color = h.color
        h.color = RED
        return x

    def _flip_colors(self, h):
        h.color = RED
        h.left.color = BLACK
        h.right.color = BLACK

    def _insert(self, h, element):
        if h is None:
            return LLRBNode(element, RED), True

        inserted = False
        if element < h.key:
            h.left, inserted = self._insert(h.left, element)
        elif element > h.key:
            h.right, inserted = self._insert(h.right, element)
        else:
            return h, False

        if _is_red(h.right) and not _is_red(h.left):
            h = self._rotate_left(h)
        if _is_red(h.left) and _is_red(h.left.left):
            h = self._rotate_right(h)
        if _is_red(h.left) and _is_red(h.right):
            self._flip_colors(h)

        return h, inserted

    def insertElement(self, element):
        inserted = False

        self.root, inserted = self._insert(self.root, element)
        if self.root is not None:
            self.root.color = BLACK

        return inserted

    def searchElement(self, element):
        found = False

        current = self.root
        while current is not None:
            if element < current.key:
                current = current.left
            elif element > current.key:
                current = current.right
            else:
                found = True
                break

        return found

```

```python
import random
import string

class TestDataGenerator():
    '''
    This class generates random synthetic data based on a given set of
    characters. The default character set are digits 0-9 and the lowercase
    alphabet.
    ...

    Attributes
    ----------

    length_max : maximum length of a random string
    length_min : minimum length of a random string
    character_set : the set of characters to use

    Methods
    -------

    random_string(): returns a random string with length between `length_min`
            and `length_max` using characters from `character_set`
    random_data(count): returns `count` amount of random strings.
    sequential_data(count): returns a list of strings which "count up" to
            `count` using the characters from `character_set`. E.g.:
            sequencial_data(50) -> ["1", ... , "9", "a", "b", ... , "1e"]
    reverse_sequential_data(count): returns a list of strings counting down
            from `count`. Calls `sequential_data()` and reverses its output.

    '''

    def random_string(self):
        length = random.randint(self.__length_min, self.__length_max)
        return "".join(random.choices(self.__character_set,
                                      k=length))

    def random_data(self, count):
        return [self.random_string() for _ in range(count)]

    def sequential_data(self, count):
        output = []

        # This is just number base conversion. For number N, we convert N
        # into base a, where a is the length of the character set. We then
        # get an array of numbers representing each digit, which we then map
        # to our character set: N = 107 -> [2, 35] -> "2z"
        for i in range(count):
            digits = []
            number = i + 1
            while number:
                digits.append(number % len(self.__character_set))
                number //= len(self.__character_set)
            output.append(list(reversed(digits)))
        result = [("".join(self.__character_set[i] for i in out)) for out in output]
        return list(result)

    def reverse_sequential_data(self, count):
        result = reversed(self.sequential_data(count))
        return list(result)

    def __init__(self, length_min=20, length_max=25, character_set=
                 string.digits + string.ascii_lowercase):
        self.__character_set = character_set
        self.__length_min = length_min
        self.__length_max = length_max
        self.__character_set = character_set
```

Use the cell below to implement the requested **experimental framework** (be mindful of code readability and reusability).

```python
import timeit
import matplotlib.pyplot as plt

class ExperimentalFramework():
    '''
    A class to represent an experimental framework.

    ...

    Attributes
    ----------
    SCALES : dict
        A constant dictionary mapping scale names ("small", "medium", "large", "global") 
        to lists of dataset sizes (n) used for testing.
    data_generator : TestDataGenerator
        An instance of the data generator class used to create test datasets.
    scale_size : str
        The current testing scale selected from SCALES (e.g., "medium")

    Methods
    -------
    generate_datasets(dist: str)
        Generates a list of datasets based on a specific distribution ("random", 
        "sorted", or "reverse_sorted") according to the selected scale.

    evaluate_operation_insert(tree_class: class, data_sets: list, plot_graph: bool)
        Measures the average time taken to insert elements into a specified tree 
        type across multiple datasets.

    evaluate_operation_search(tree_class: class, data_sets: list, plot_graph: bool)
        Measures the average time taken to search for a mix of existing and 
        non-existing elements within a specified tree type.

    plot_single_graph(execution_times: list, operation: str)
        Generates a basic linear plot for a single tree's performance in 
        a specific operation.

    master_plot(tree_classes: list, data_sets: list, operation: str, log_axis: bool = False)
        Produces a comparative visualization of multiple tree structures on the 
        same axes, with an optional argument to toggle a logarithmic X-scale.

    configure_axes(ax, use_log=False)
        A utility method to toggle between linear and logarithmic scaling and 
        format the plot grid.

    '''
            
    #ADD YOUR CODE HERE
    SCALES = {
            "small": [10, 50, 100, 250, 500],
            "medium": [1000, 2500, 5000, 7500, 10000],
            "large":[25000, 50000, 100000, 250000, 500000],
            "global":[10, 100, 1000, 10000, 100000, 500000]
        }
    
    def __init__(self, scale_size):
        self.data_generator = TestDataGenerator()
        self.scale_size = scale_size

    
    def generate_datasets(self, dist: str):
        distributions = ["random", "sorted", "reverse_sorted"]
        
        scale_size = self.scale_size
        if dist not in distributions:
            pass
        else:
            data_sets = list()
            scale = self.SCALES.get(scale_size)
            # Generate a list of datasets to test on depending on (1) the distribution and (2) the scale of the size of data sets chosen
            if dist == "random":
                for n in range(len(scale)):
                    data_set = self.data_generator.random_data(scale[n])
                    data_sets.append(data_set)
            elif dist == "sorted":
                for n in range(len(scale)):
                    data_set = self.data_generator.sequential_data(scale[n])
                    data_sets.append(data_set)
            else:
                for n in range(len(scale)):
                    data_set = self.data_generator.reverse_sequential_data(scale[n])
                    data_sets.append(data_set)

            return data_sets

    def evaluate_operation_insert(self, tree_class, data_sets: list, plot_graph: bool):
        execution_times = []
        # n_insert = number of insertions per dataset. Reduce amount for scales with larger datasets
        n_insert = 10 if self.scale_size=="large" or self.scale_size=="global" else 100 
        
        for data_set in data_sets:
            data_structure = tree_class()
            for i in data_set:
                data_structure.insertElement(i)
                
            test_elements = self.data_generator.random_data(n_insert)
            
            samples = timeit.repeat(
                stmt='for e in elements: data_structure.insertElement(e)',
                globals={'data_structure': data_structure, 'elements': test_elements},
                number=1,
                repeat=20
            )
            
            execution_times.append(min(samples) / n_insert) # Divide by n to get the time for one insertion
        if plot_graph:
            self.plot_single_graph(execution_times, "Insert")
        return execution_times
    
    def evaluate_operation_search(self, tree_class, data_sets: list, plot_graph: bool):
        execution_times = []
        n_search = 100 # number of searches per dataset
        
        for data_set in data_sets:
            data_structure = tree_class()
            for i in data_set:
                data_structure.insertElement(i)
        
            # Half of testing set will consist of "in" elements that are within the set and the other half, "out" elements which are not. 
            in_elements = data_set[:n_search//2] 
            out_elements = self.data_generator.random_data(n_search//2)
            test_elements = in_elements + out_elements
            
            samples = timeit.repeat(
                stmt='for e in elements: data_structure.searchElement(e)',
                globals={'data_structure': data_structure, 'elements': test_elements},
                number=1,
                repeat=20
            )
            
            execution_times.append(min(samples) / n_search)

        if plot_graph:
            self.plot_single_graph(execution_times, "Search")

        return execution_times
    
    def plot_single_graph(self, execution_times: list, operation: str):
        x = self.SCALES.get(self.scale_size)

        plt.figure(figsize=(10, 6))
        plt.plot(x, execution_times, marker='o', linestyle='-', color='b')
    
        plt.xlabel('Size of dataset (n)')
        plt.ylabel('Time (seconds)')
        plt.grid(True)
        plt.title(f'{operation} Performance: ({self.scale_size} scale)')
        plt.show()

        print(execution_times)

    def master_plot(self, tree_classes, data_sets: list, operation: str, log_axis: bool = False):
        # First create figure
        fig, ax = plt.subplots(figsize=(10, 6))
        x = self.SCALES.get(self.scale_size)
        colours = ["green", "blue", "red"]

        # Evaluate execution times for each tree and add to plot
        for i, tree_class in enumerate(tree_classes):
            if operation == "Insert":
                execution_times = self.evaluate_operation_insert(tree_class, data_sets, False)
            elif operation == "Search": 
                execution_times = self.evaluate_operation_search(tree_class, data_sets, False)
            else:
                raise ValueError("Please pass in operation argument as \"Search\" or \"Insert\"")
            ax.plot(x, execution_times, marker='o', label=type(tree_class()).__name__, color=colours[i])
        
        self.configure_axes(ax, use_log=log_axis)

        ax.set_title(f'Master plot of {operation} Performance: ({self.scale_size} scale)')
        ax.set_xlabel('Size of dataset (n)')
        ax.set_ylabel('Time per Operation (seconds)')
        ax.legend()
        ax.grid(True, linestyle='--', alpha=0.7)

        plt.show()  

    def configure_axes(self, ax, use_log=False):
        if use_log:
            ax.set_xscale('log')
            # Adding minor grids makes log scales much easier to interpret
            ax.grid(True, which='both', linestyle='--', alpha=0.5)
        else:
            ax.set_xscale('linear')
            ax.grid(True, which='major', linestyle='--', alpha=0.7)  
    
    
    
    
```
